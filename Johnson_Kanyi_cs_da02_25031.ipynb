```python
# Importing libraries for web scraping
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Set URL of the website
url = 'https://www.scrapethissite.com/pages/forms/'
page = requests.get(url)

# Parse HTML content with BeautifulSoup
soup = BeautifulSoup(page.text, 'html.parser')

# Print to confirm we fetched the HTML
print(soup.prettify()[:500])  # Show first 500 characters
```

```python
# Extract the hockey scores table
hockey_table = soup.find('table', class_='table')
print(hockey_table)
```

```python
# Extract the column headings
table_titles = hockey_table.find_all('th')
hockey_table_title = [title.text.strip() for title in table_titles]
print(hockey_table_title)
```

```python
# Create empty DataFrame with headings
df = pd.DataFrame(columns=hockey_table_title)
df.head()
```

```python
# Extract all rows
table_data = hockey_table.find_all('tr')

# Loop through rows and extract data
for row in table_data[1:]:
    raw_data = row.find_all('td')
    each_raw_data = [data.text.strip() for data in raw_data]

    # Add row into DataFrame
    length = len(df)
    df.loc[length] = each_raw_data

# Inspect DataFrame
df.head()
```

```python
# Save DataFrame to CSV
df.to_csv('Hockey.csv', index=False)

print("Data successfully saved as Hockey.csv")
```
